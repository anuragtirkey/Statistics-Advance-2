1. What is hypothesis testing in statistics?

Ans.  Hypothesis testing is a method in statistics used to decide whether there is enough evidence to support a specific claim, called a hypothesis about a population. It compares data from a sample to what is expected under the claim and determines if the differences are significant or just due to chance.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans. The **null hypothesis is a statement that assumes no effect, no difference, or no relationship exists in the data. It is the default assumption.

The alternative hypothesis is the opposite—it suggests there is an effect, difference, or relationship.

  For example:
  - Null Hypothesis: A new medicine has no effect.
  - Alternative Hypothesis: The new medicine works.

In hypothesis testing, we aim to see if the data provides enough evidence to reject the null hypothesis in favor of the alternative.

3. What is the significance level in hypothesis testing, and why is it important?

Ans. The significance level is the threshold we set to decide whether to reject the null hypothesis. Common values are 0.05 or 0.01.
It tells us the maximum chance we're willing to accept of making a mistake by rejecting the null hypothesis when it is actually true.

For example:

  alpha = 0.05, it means we're allowing a 5% chance of rejecting the null hypothesis incorrectly.

It's important because it helps control how strict or lenient we are in drawing conclusions from data.

4. What does a P-value represent in hypothesis testing?

Ans. The P-value is the probability of observing the data if the null hypothesis is true.

It helps us decide whether to reject the null hypothesis:

  - A small P-value (e.g. < 0.05) means the data is unlikely under the null hypothesis, so we reject it.
  - A large P-value means the data is consistent with the null hypothesis, so we don't reject it.

It basically shows how strong the evidence is against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?

Ans. To interpret the P-value, this is what we keep in mind:

  - If the P-value is small (e.g. less than the significance level, like 0.05): This means the observed data is very unlikely under the null hypothesis. So, we reject the null hypothesis and conclude there is evidence for the alternative hypothesis.

  - If the P-value is large (e.g. greater than 0.05): This means the data is consistent with the null hypothesis. So, we fail to reject the null hypothesis and cannot claim evidence for the alternative hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Ans. Type 1 and Type 2 errors are mistakes we can make in hypothesis testing:

  - Type 1 Error: Rejecting the null hypothesis when it is actually true. (A "false positive" — like thinking a medicine works when it doesn't.)
  
  - Type 2 Error: Failing to reject the null hypothesis when it is actually false. (A "false negative" — like thinking a medicine doesn't work when it actually does.)

These errors highlight the balance between being cautious (avoiding Type 1 errors) and being sensitive (avoiding Type 2 errors).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans. A one-tailed test checks for an effect or difference in only one direction (either greater than or less than a certain value). It's used when you're only interested in one specific outcome.

A two-tailed test checks for an effect or difference in both directions (greater than or less than a certain value). It's used when you're interested in any significant change, regardless of the direction.

For example:

  - One-tailed: Is this medicine better than the current one?
  - Two-tailed: Does this medicine work differently (better or worse) than the current one?

8. What is the Z-test, and when is it used in hypothesis testing?

Ans. The Z-test is a statistical test used to compare sample data to a population when:

  1. The data is normally distributed.
  2. The population variance is known.
  3. The sample size is large (generally, n > 30).

It helps determine if there's a significant difference between the sample mean and the population mean, or between two sample means. It's often used in cases like testing proportions or averages.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans. The Z-score measures how far a data point or sample statistic is from the population mean, in terms of standard deviations.

We calculate it using this formula: Z = (X - mu) / sigma

Where:
  - X is the observed value or sample statistic.
  - mu is the population mean.
  - sigma is the population standard deviation.

In hypothesis testing, the Z-score helps determine whether the sample data is significantly different from the population. A higher absolute Z-score (far from 0) suggests stronger evidence against the null hypothesis.

20. What is an ANOVA test, and what are its assumptions?

22. What is the F-test, and how does it relate to hypothesis testing?

Ans. The F-test is a statistical test used to compare the variances of two or more groups to determine if they are significantly different. It plays a crucial role in hypothesis testing, especially in contexts like ANOVA and regression analysis.

How it relates to hypothesis testing:

  1. Set up the hypotheses:

   - Null hypothesis: Variances are equal (no significant difference).
   - Alternative hypothesis: Variances are not equal.

  2. Calculate the F-statistic
   
  3. Compare with the critical value:
  
   Based on the F-distribution, degrees of freedom, and significance level (e.g. 0.05), the calculated F-statistic is compared to a critical value.

  4. Decision:
  
   - If the F-statistic > critical value: Reject the null hypothesis (variances are significantly different).
   - If the F-statistic <= critical value: Fail to reject the null hypothesis (variances are similar).

The F-test is widely used in ANOVA and regression to check whether variability between groups or models is meaningful.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, significance_level=0.05):

    z_stat = (sample_mean - population_mean) / (population_std / (sample_size ** 0.5))


    critical_value = norm.ppf(1 - significance_level / 2)

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print("Z-statistic:", z_stat)
    print("Critical value:", critical_value)
    print("P-value:", p_value)

    if abs(z_stat) > critical_value:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# Example
sample_mean = 78
population_mean = 75
population_std = 10
sample_size = 40
significance_level = 0.05

# Perform the Z-test
z_test(sample_mean, population_mean, population_std, sample_size, significance_level)
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std):

    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    return z_stat, p_value

# Example
sample = [78, 75, 82, 79, 74, 77, 76, 80, 85, 73]
population_mean = 75
population_std = 5

z_stat, p_value = one_sample_z_test(sample, population_mean, population_std)

print("Z-statistic:", z_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha, beta, mu0, mu1, sigma, n):

    critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma / np.sqrt(n))

    x = np.linspace(mu0 - 4 * sigma, mu1 + 4 * sigma, 1000)

    h0 = norm.pdf(x, loc=mu0, scale=sigma / np.sqrt(n))

    h1 = norm.pdf(x, loc=mu1, scale=sigma / np.sqrt(n))

    plt.figure(figsize=(10, 6))
    plt.plot(x, h0, label="H0: Null Hypothesis", color="blue")
    plt.plot(x, h1, label="H1: Alternative Hypothesis", color="green")

    plt.fill_between(x, h0, where=(x >= critical_value), color="red", alpha=0.5, label="Type 1 Error (α)")

    plt.fill_between(x, h1, where=(x < critical_value), color="orange", alpha=0.5, label="Type 2 Error (β)")

    plt.axvline(critical_value, color="black", linestyle="--", label=f"Critical Value = {critical_value:.2f}")

    plt.title("Visualization of Type 1 and Type 2 Errors")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example
alpha = 0.05
beta = 0.2
mu0 = 100
mu1 = 110
sigma = 15
n = 30

visualize_errors(alpha, beta, mu0, mu1, sigma, n)
```

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import ttest_ind, norm

def z_test(group1, group2):

    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    n1, n2 = len(group1), len(group2)

    pooled_std = np.sqrt((std1**2 / n1) + (std2**2 / n2))

    z_stat = (mean1 - mean2) / pooled_std

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    return z_stat, p_value

def compare_tests():

    np.random.seed(42)
    group1 = np.random.normal(loc=50, scale=10, size=30)
    group2 = np.random.normal(loc=55, scale=10, size=30)

    t_stat, t_p_value = ttest_ind(group1, group2)

    z_stat, z_p_value = z_test(group1, group2)

    print("T-test results:")
    print("T-statistic:", t_stat)
    print("P-value:", t_p_value)

    print("\nZ-test results:")
    print("Z-statistic:", z_stat)
    print("P-value:", z_p_value)

    print("\nComparison:")
    if np.isclose(t_p_value, z_p_value, atol=1e-2):
        print("The T-test and Z-test give similar results.")
    else:
        print("The T-test and Z-test give different results.")

compare_tests()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(sample, confidence_level=0.95):

    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)

    z = norm.ppf((1 + confidence_level) / 2)

    margin_of_error = z * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example
sample_data = [78, 75, 82, 79, 74, 77, 76, 80, 85, 73]
confidence = 0.95
interval = confidence_interval(sample_data, confidence)
print(f"95% Confidence Interval: {interval}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior_A, likelihood_B_given_A, likelihood_B_given_not_A, prior_not_A):

    marginal_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)

    posterior_A_given_B = (likelihood_B_given_A * prior_A) / marginal_B

    return posterior_A_given_B

# Example
prior_A = 0.3
likelihood_B_given_A = 0.8
likelihood_B_given_not_A = 0.4
prior_not_A = 1 - prior_A

posterior = bayesian_inference(prior_A, likelihood_B_given_A, likelihood_B_given_not_A, prior_not_A)
print(f"Posterior probability P(A|B): {posterior:.2f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np

def calculate_expected_frequencies(observed_data):

    observed_data = np.array(observed_data)

    row_sums = np.sum(observed_data, axis=1)
    column_sums = np.sum(observed_data, axis=0)
    total = np.sum(observed_data)

    expected_frequencies = np.zeros_like(observed_data, dtype=float)
    for i in range(observed_data.shape[0]):
        for j in range(observed_data.shape[1]):
            expected_frequencies[i, j] = (row_sums[i] * column_sums[j]) / total

    return expected_frequencies

# Example
observed_data = [
    [50, 30, 20],
    [30, 40, 30],
    [20, 30, 50],
]

expected = calculate_expected_frequencies(observed_data)
print("Observed Data:\n", np.array(observed_data))
print("Expected Frequencies:\n", expected)
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):

    chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

    print("Chi-square Statistic:", chi2_stat)
    print("P-value:", p_value)

    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

# Example
observed_data = [50, 30, 20]
expected_data = [40, 40, 20]

goodness_of_fit_test(observed_data, expected_data)
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * min(1 - f.cdf(f_stat, df1, df2), f.cdf(f_stat, df1, df2))

    print("F-statistic:", f_stat)
    print("P-value:", p_value)

    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Example
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=15, size=30)

f_test(sample1, sample2)
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt

def check_anova_assumptions(groups):

    print("1. Checking Normality (Shapiro-Wilk Test):")
    for i, group in enumerate(groups, start=1):
        stat, p_value = shapiro(group)
        print(f"Group {i}: W-statistic={stat:.4f}, p-value={p_value:.4f}")
        if p_value < 0.05:
            print(f"   Group {i} violates normality assumption (p-value < 0.05).")
        else:
            print(f"   Group {i} satisfies normality assumption (p-value >= 0.05).")

    print("\n2. Checking Equal Variances (Levene's Test):")
    stat, p_value = levene(*groups)
    print(f"Levene's Test: W-statistic={stat:.4f}, p-value={p_value:.4f}")
    if p_value < 0.05:
        print("   Variances are significantly different (p-value < 0.05).")
    else:
        print("   Variances are equal (p-value >= 0.05).")

    print("\n3. Checking Independence:")
    print("   Independence is a design assumption, and should be ensured during data collection.")

    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups, start=1):
        plt.hist(group, alpha=0.6, bins=10, label=f"Group {i}", density=True)
    plt.title("Distribution of Groups (Check Normality)")
    plt.xlabel("Values")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=55, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)

groups = [group1, group2, group3]

check_anova_assumptions(groups)
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2

def test_population_variance(sample, hypothesized_variance, alpha=0.05):

    sample_size = len(sample)
    sample_variance = np.var(sample, ddof=1)
    df = sample_size - 1  # Degrees of freedom

    chi_square_stat = (df * sample_variance) / hypothesized_variance

    chi2_critical_low = chi2.ppf(alpha / 2, df)
    chi2_critical_high = chi2.ppf(1 - alpha / 2, df)

    print("Chi-square Statistic:", chi_square_stat)
    print(f"Critical Values: Lower = {chi2_critical_low:.4f}, Upper = {chi2_critical_high:.4f}")

    if chi_square_stat < chi2_critical_low or chi_square_stat > chi2_critical_high:
        print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in the population variance.")

# Example
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
hypothesized_var = 100
alpha = 0.05

test_population_variance(sample_data, hypothesized_var, alpha)
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Function to perform F-test
def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    critical_value = f.ppf(1 - alpha, df1, df2)
    p_value = 2 * min(1 - f.cdf(f_stat, df1, df2), f.cdf(f_stat, df1, df2))

    print("F-statistic:", f_stat)
    print("Degrees of Freedom: df1 =", df1, ", df2 =", df2)
    print("Critical F-value:", critical_value)
    print("P-value:", p_value)

    if f_stat > critical_value:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, df1, df2)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color="blue", label=f"F-distribution (df1={df1}, df2={df2})")
    plt.axvline(f_stat, color="red", linestyle="--", label=f"F-statistic = {f_stat:.2f}")
    plt.axvline(critical_value, color="green", linestyle="--", label=f"Critical Value = {critical_value:.2f}")
    plt.fill_between(x, y, where=(x >= critical_value), color="gray", alpha=0.3, label="Rejection Region (alpha=0.05)")
    plt.title("F-test Visualization")
    plt.xlabel("F-statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=15, size=30)

f_test(sample1, sample2)
```
